```python
class Patient:
    def __init__(self,id,name,age,blood_g):
        self.id=id
        self.name=name
        self.age=age
        self.group=blood_g
        self.next,self.prev=None,None
```

```python
class WRM :
    def __init__(self):
        self.d_head=Patient(None,None,None,None)
        self.d_head.prev=self.d_head
        self.d_head.next=self.d_head
        self.d_tail=self.d_head

    def RegisterPatient(self,id=None,name=None,age=None,blood=None):

        if id !=None and name !=None and age !=None and blood !=None :
            data=Patient(id,name,age,blood)
            data.prev=self.d_tail
            self.d_tail.next=data
            self.d_tail=self.d_tail.next
            self.d_head.prev=self.d_tail
            data.next=self.d_head
            print(f"Patient {name} has been succesfully registered !")
        else:
            print("---> Sorry, Patiens full information is required to register ! <----")
        print("_"*80,"\n")

    def Servepatient(self):
        print("="*42)
        if self.d_head.next==self.d_head:
            print("Sorry there are no more patiensts to serve !")
        else:
            data=self.d_head.next
            self.d_head.next.next.prev=self.d_head
            self.d_head.next=self.d_head.next.next
            data.next=None
            data.prev=None
            print(f"|| The patient {data.name} has been Treated ! ||")
        print("="*42,"\n")

    def CancelAll(self):
        print("="*60)
        if self.d_head.next==self.d_head:
            print("Unsuccessful ! There are no patients. Nothing to cancel ! ")
        else:
            self.d_head.next=self.d_head
            self.d_head.prev=self.d_head
            print("Sucessfull !! All appointments has been Cancelled !")
        print("="*60,"\n")
    def CanDoctorGoHome(self):
        print("="*30)
        if self.d_head.next==self.d_head:
            print("Yes! the Doctor can Go home now ! ")
        else:
            print("No! the Doctor can not go home ! ")
        print("="*30,"\n")

    def ShowAllPatient(self):
        print("="*35)
        temp=self.d_head.next
        print("||",end=" ")
        if self.d_head.next.id==None:
            print("There are No patients Waiting  ",end="")
        while temp!=self.d_head:
            print(temp.id,end=" ")
            temp=temp.next
        print("||")
        print("="*35,"\n")



obj=WRM()
```

```python
#dashboard type driver code
def dash(data):
    if data==1:
        id=int(input("Enter patient Id:"))
        name=str(input("Enter patient Name:"))
        age=int(input("Enter patient Age:"))
        blood=str(input("Enter patient Blood Group:"))
        print("\nWelcome to Bracu Hospital !\n")
        if name!="" and blood!="":
            obj.RegisterPatient(id,name,age,blood)
        else:
            print("Sorry all fields must be completed to register a patient !")

    elif data==2:
        obj.Servepatient()

    elif data==3:
        obj.CancelAll()

    elif data==4:
        obj.CanDoctorGoHome()

    elif data==5:
        obj.ShowAllPatient()
    else:

        print('Invalid Operation')

print("=======Hospital Dashboard=======")
print(f"----> Enter your choise <----\n\n1.Register Patient\n2.Serve Patient\n3.Cancell All\n4.Can DoctorGo Home\n5.Show All Patient")
print("================================")
temp=int(input("Enter Your option No."))
dash(temp)
```
