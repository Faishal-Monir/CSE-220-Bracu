```python
import numpy as np         #You need to write this code !
class Node:
    def __init__(self,elem,next=None):
        self.elem=elem
        self.next=None
```

```python
def printLinkedList(head): #Will be given
  temp = head
  while temp != None:
    if temp.next != None:
      print(temp.elem, end = '-->')
    else:
      print(temp.elem)
    temp = temp.next
  print()
```

```python
def createList(arr):    #You need to write this code !
    head=Node(arr[0],None)
    tail=head
    for i in range(1,len(arr)):
        temp=Node(arr[i],None)
        tail.next=temp
        tail=tail.next
    return head
```

```python
def remove_dup(head): #You need to write this code !
    n_head=None
    n_tail=None
    tail=head
    while tail!=None:
        if n_head==None:
            n_head=Node(tail.elem,None)
            n_tail=n_head
        elif n_tail.elem==tail.elem:
            pass
        elif n_tail.elem!=tail.elem:
            temp=Node(tail.elem,None)
            n_tail.next=temp
            n_tail=temp
        tail=tail.next
    return n_head
```

```python
#Tester Code #Will be given
head=createList(np.array([2,2,2,2,7,7,9,10,10,10,10,12]))
printLinkedList(head) #This should print    2-->2-->2-->2-->7-->7-->9-->10-->10-->10-->10-->12
h=remove_dup(head)
printLinkedList(h)    #This should print    2-->7-->9-->10-->12
```
