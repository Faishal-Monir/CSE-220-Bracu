```python
#You must run this cell to print matrix and for the driver code to work
def print_matrix(m):
  row,col = m.shape
  for i in range(row):
    c = 1
    print('|', end='')
    for j in range(col):
      c += 1
      if(len(str(m[i][j])) == 1):
        print(' ',m[i][j], end = '  |')
        c += 6
      else:
        print(' ',m[i][j], end = ' |')
        c += 6
    print()
    print('-'*(c-col))

```

Task 1: Zigzag Walk

```python
def walk_zigzag(floor):
  row,col=floor.shape
  for i in range(0,row):
    if i%2==0:
        for j in range(0,col,2):
            print(floor[i][j],end=" ")
        print()
    else:
         for k in range(col-1,-1,-2):
             print(floor[i][k],end=" ")
         print()




floor = np.array([[ '3W' , '8B' , '4W' , '6B' , '1W' , '5B'],['3B' , '2W' , '1B' , '6W' , '3B' , '8W'],
                  ['9W' , '0B' , '7W' , '5B' , '3W' , '8B'],['2B' , '1W' , '3B' , '6W' , '0B' , '4W'],
                   ['1W' , '4B' , '2W' , '8B' , '6W' , '6B']])
print_matrix(floor)
print('Walking Sequence:')
walk_zigzag(floor)
#This should print
# 3W 4W 1W
# 8W 6W 2W
# 9W 7W 3W
# 4W 6W 1W
# 1W 2W 6W
```

Task 2: Landscape Screen

```python
def landscape(m):
    row,col=m.shape
    mat=np.zeros((col,row),dtype=np.int64)
    for i in range(0,row,1):
        for j in range(0,col,1):
            mat[j][i]=m[i][j]
    return mat


m = np.random.randint(15,size=(4,3)) #This creates an np array of 4 x 3 where the values are random numbers from [0,15] range
print_matrix(m)
print()
print_matrix(landscape(m))
```

Task 3: Seating Arrangement

```python
import numpy as np
def arrangement(m, diff):
  row,col=m.shape
  for i in range(row):
    for j in range(col):
        if i==j:
            pass
        elif j>i:                #Only taking the upper half as the results would be same for the lower half
            x=m[i][j]-m[j][i]
            if x<0:              #makink the absolute value or working with negative numbers
                x=-x
            if x!=diff:
                return False
  return True
#=======================================================================
m = np.array([[7,13,9,14],[12,8,15,11],[10,17,3,2],[15,10,1,4]])
print_matrix(m)
print()
returned_value = arrangement(m,1)
print(returned_value)#This should print False
unittest.output_test(returned_value, False)
print("===================================><===========================\n")
m = np.array([[8,15,7,12],[13,2,18,11],[9,20,5,2],[14,9,0,10]])
print_matrix(m)
print()
returned_value = arrangement(m,2)
print(returned_value)#This should print True
unittest.output_test(returned_value, True)
```

Task 4: Chess Piece

```python
def show_knight_move(knight):
    chess=np.zeros((8,8),dtype=np.int64)
    row,col=knight
    if 0<=row<=7 and 0<=col<=7:
        chess[row][col]=33
    else:
        return "Not possible for chess Game!"

    if 0<=row-2<=7:                   # For upward direction
        if 0<=col-1<=7:
            chess[row-2][col-1]=3
        if 0<=col+1<=7:
            chess[row-2][col+1]=3
    else:
        pass


    if 0<=row+2<=7:                    # For Downward direction
        if 0<=col-1<=7:
            chess[row+2][col-1]=3
        if 0<=col+1<=7:
            chess[row+2][col+1]=3
    else:
        pass


    if 0<=col-2<=7:                    # For leftward direction
        if 0<=row-1<=7:
            chess[row-1][col-2]=3
        if 0<=row+1<=7:
            chess[row+1][col-2]=3
    else:
        pass


    if 0<=col+2<=7:                    # For rightward direction
        if 0<=row-1<=7:
            chess[row-1][col+2]=3
        if 0<=row+1<=7:
            chess[row+1][col+2]=3
    else:
        pass



    return chess


knight = (3,4)
chess_board = show_knight_move(knight)
print_matrix(chess_board)
```

```python
def topple(knight,rook):
    k_row,k_col=knight
    r_row,r_col=rook
    chess=np.zeros((8,8),dtype=np.int64)

#_____________________Initial condition check _________________________
    if 0<=k_row<=7 and 0<=k_col<=7:
        chess[k_row][k_col]=33
    else:
        return "Not possible for chess Game!"

    if 0<=r_row<=7 and 0<=r_col<=7:
        chess[r_row][r_col]=5
    else:
        return "Not possible for chess Game!"
#__________________________For-Knight__________________________________
    status="None can kill !"

    if 0<=k_row-2<=7:                   # For upward direction
        if 0<=k_col-1<=7:
            chess[k_row-2][k_col-1]=3
            if chess[k_row-2][k_col-1]==chess[r_row][r_col]:
                status="Knight can kill !"

        if 0<=k_col+1<=7:
            chess[k_row-2][k_col+1]=3
            if chess[k_row-2][k_col+1]==chess[r_row][r_col]:
                status="Knight can kill !"
    else:
        pass
#___________________________________________________________________________

    if 0<=k_row+2<=7:                    # For Downward direction
        if 0<=k_col-1<=7:
            chess[k_row+2][k_col-1]=3
            if chess[k_row+2][k_col-1]==chess[r_row][r_col]:
                status="Knight can kill !"
        if 0<=k_col+1<=7:
            chess[k_row+2][k_col+1]=3
            if chess[k_row+2][k_col+1]==chess[r_row][r_col]:
                status="Knight can kill !"
    else:
        pass
#___________________________________________________________________________

    if 0<=k_col-2<=7:                    # For leftward direction
        if 0<=k_row-1<=7:
            chess[k_row-1][k_col-2]=3
            if chess[k_row-1][k_col-2]==chess[r_row][r_col]:
                status="Knight can kill !"
        if 0<=k_row+1<=7:
            chess[k_row+1][k_col-2]=3
            if chess[k_row+1][k_col-2]==chess[r_row][r_col]:
                status="Knight can kill !"
    else:
        pass
#___________________________________________________________________________

    if 0<=k_col+2<=7:                    # For rightward direction
        if 0<=k_row-1<=7:
            chess[k_row-1][k_col+2]=3
            if chess[k_row-1][k_col+2]==chess[r_row][r_col]:
                status="Knight can kill !"
        if 0<=k_row+1<=7:
            chess[k_row+1][k_col+2]=3
            if chess[k_row+1][k_col+2]==chess[r_row][r_col]:
                status="Knight can kill !"
    else:
        pass
#______________________________________________________________________________
#__________________________Now we check for rook_______________________________

    for i in range(0,8):                         #horizontal rook itteration
        if chess[r_row][i]==chess[k_row][k_col]:
            status="Rook can kill ! "
    for i in range(0,8):                         #vertical rook itteration
        if chess[i][r_col]==chess[k_row][k_col]:
            status="Rook can kill !"

    return chess,status


knight = (5,6)
rook = (5,1)
chess_board,ans= topple(knight,rook) #,ans
print_matrix(chess_board)
print(ans) #This should print Rook can kill
```
